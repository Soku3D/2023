```python
import matplotlib.pyplot as plt
```

```python
data = {  'apple ': 10, 'lemon':5}
names = list(data.keys())
values= list(data.values())
flag, axs = plt.subplots(1,3,figsize=(12,10), sharey=True)
axs[0].bar(names, values,color='yellow')
axs[1].scatter(names,values)
axs[2].plot(names,values, color='black')
flag.suptitle('plot')
```
